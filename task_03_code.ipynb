```python
import pandas as pd 
```

***Đọc dữ liệu***

```python
shipper = pd.read_csv("shipper_info_clean.csv")
order = pd.read_csv("order_info_clean.csv")
rating = pd.read_csv("customer_rating_clean.csv")
```

***Thời gian trung bình theo shipper***

```python
avg_time_shipper = (
    order
    .groupby("shipper_id")["delivery_time_min"]
    .mean()
    .reset_index(name="avg_delivery_time")
)

avg_time_shipper
```

***Quãng đường trung bình theo từng khu vực***


```python
order_shipper = order.merge(
    shipper[["shipper_id", "region"]],
    on="shipper_id",
    how="left"
)

avg_distance_region = (
    order_shipper
    .groupby("region")["distance_km"]
    .mean()
    .reset_index(name="avg_distance_km")
)

avg_distance_region
```

***Xác định shipper giao hàng nhanh nhất***

```python
fastest_shipper = avg_time_shipper.sort_values(
    by="avg_delivery_time"
).head(1)

fastest_shipper
```

***Phân tích mối liên hệ giữa rating và delivery_time***

```python
order_rating = order.merge(
    rating,
    on="order_id",
    how="left"
)
rating_vs_time = (
    order_rating
    .groupby("rating")["delivery_time_min"]
    .mean()
    .reset_index(name="avg_delivery_time")
)

rating_vs_time
```

***Tìm nhóm shipper có hiệu suất thấp***


```python
overall_avg_time = order["delivery_time_min"].mean()
overall_avg_time
low_performance_shipper = avg_time_shipper[
    avg_time_shipper["avg_delivery_time"] > overall_avg_time
]

low_performance_shipper
```

***Xuất ra kết quả***

```python
avg_time_shipper.to_csv("avg_delivery_time_by_shipper.csv", index=False)
avg_distance_region.to_csv("avg_distance_by_region.csv", index=False)
low_performance_shipper.to_csv("low_performance_shipper.csv", index=False)
```
