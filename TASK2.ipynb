```python
import pandas as pd

# Đọc dữ liệu từ các file CSV
shipper_df = pd.read_csv("shipper_info_clean.csv")
rating_df = pd.read_csv("customer_rating_clean.csv")
order_df = pd.read_csv("order_info_clean.csv")

shipper_df.head(), rating_df.head(), order_df.head()
```

```python
# Lọc shipper theo từng khu vực
print("=== Shipper theo khu vực ===")
for region in shipper_df['region'].unique():
    shippers = shipper_df[shipper_df['region'] == region]['full_name'].unique()
    print(f"Khu vực {region}: {list(shippers)}")
```

```python
# Thống kê số đơn hàng theo shipper
orders_per_shipper = order_df['shipper_id'].value_counts()
print("=== Số đơn hàng theo shipper ===")
print(orders_per_shipper)
```

```python
# Tìm đơn hàng không có rating
orders_with_rating = rating_df['order_id'].unique()
orders_without_rating = order_df[~order_df['order_id'].isin(orders_with_rating)]

print("=== Đơn hàng không có rating ===")
print(orders_without_rating[['order_id', 'shipper_id']])
print(f"Tổng số đơn hàng không có rating: {len(orders_without_rating)}")
```

```python
# Tính quãng đường trung bình
avg_distance = order_df['distance_km'].dropna().mean()
print(f"=== Quãng đường trung bình: {avg_distance:.2f} km ===")
```

```python
# Ghép dữ liệu order với shipper để biết khu vực
merged_df = order_df.merge(shipper_df, on="shipper_id", how="left")

# Tính trung bình thời gian giao hàng theo khu vực
avg_delivery_by_region = merged_df.groupby("region")['delivery_time_min'].mean()
print("=== Thời gian giao hàng trung bình theo khu vực ===")
print(avg_delivery_by_region)   
```
